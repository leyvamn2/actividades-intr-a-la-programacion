```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
import urllib.request
from PIL import Image
```

```python
urlPicture="https://upload.wikimedia.org/wikipedia/commons/thumb/8/82/Circle-yellow.svg/1200px-Circle-yellow.svg.png"
internetPicture=Image.open(urllib.request.urlopen(urlPicture))
# la imagen se obtiene y se carga en una variable, despu√©s se convierte en matriz
yellowCircle=np.array(internetPicture)
```

```python
plt.imshow(yellowCircle)
```

```python
#nueva variable
faceTest=np.array(yellowCircle)
plt.figure(figsize=(10,20))
plt.imshow(faceTest)
```
