# Indicadores del Banco Mundial
## PIB vs Calidad de Vida

```python
import pandas as pd
import numpy as np
import random #para generar numeros aleatorios
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('Indicators.csv')
```

### La primer información del indicador que se va a mostrar va a ser del PIB per cápita de ese país

```python
hist_indicador= 'GDP per capita \(constant 2005'
mask1 = data['IndicatorName'].str.contains(hist_indicador)
hist_country = 'TUR'
mask2 = data['CountryCode'].str.contains(hist_country)
gdp_stage = data[mask1 & mask2]
gdp_stage.head()
```

#### Se muestra gráfica del PIB a través de los años

```python

plt.plot(gdp_stage['Year'],gdp_stage['Value'])
plt.title('PIB Turquia')
plt.xlabel('Año')
plt.ylabel('GDP')
plt.show()

print('Primer año de registro',min(gdp_stage['Year']))
print('Último año de registro',max(gdp_stage['Year']))
```

### La segunda información que se analizará es el crecimiento de la población desde 1960 hasta 2014

```python
hist_indicador= 'Population, total'
mask1 = data['IndicatorName'].str.contains(hist_indicador)
hist_country = 'TUR'
mask2 = data['CountryCode'].str.contains(hist_country)
population_stage = data[mask1 & mask2]
population_stage.head()
```

```python

plt.plot(population_stage['Year'],population_stage['Value'])
plt.title('Población de Turquía desde 1960 hasta 2014')
plt.xlabel('Año')
plt.ylabel('Personas')
plt.show()
print('Primer año de registro',min(population_stage['Year']))
print('Último año de registro',max(population_stage['Year']))
```

## Relación entre PIB y Crecimiento de la Población en Turquía desde 1960 hasta 2014

```python
fig, axis = plt.subplots()
axis.yaxis.grid(True)
axis.set_title('Relacion de PIB y Crecimiento Poblacional de 1960 hasta 2014  en Turquía')
axis.set_xlabel(gdp_stage['IndicatorName'].iloc[0], fontsize = 10)
axis.set_ylabel(population_stage['IndicatorName'].iloc[0], fontsize = 10)
X = gdp_stage['Value']

Y = population_stage['Value']
axis.scatter(X,Y)
plt.show()

```

```python
#Urban population
hist_indicador= 'Industry, value added \(current US'
mask1 = data['IndicatorName'].str.contains(hist_indicador)
hist_country = 'TUR'
mask2 = data['CountryCode'].str.contains(hist_country)
industry_value = data[mask1 & mask2]
industry_value.head()

```

```python
plt.plot(industry_value['Year'],industry_value['Value'])
plt.title('Valor añadido de la industria($USD) desde 1960 hasta 2014')
plt.xlabel('Año')
plt.ylabel('%Valor')
plt.show()
print('Primer año de registro',min(industry_value['Year']))
print('Último año de registro',max(industry_value['Year']))
```

```python
fig, axis = plt.subplots()
axis.yaxis.grid(True)
axis.set_title('Relacion de PIB y Valor añadido de la industria desde 1960 hasta 2014  en Turquía')
axis.set_xlabel(gdp_stage['IndicatorName'].iloc[0], fontsize = 10)
axis.set_ylabel(industry_value['IndicatorName'].iloc[0], fontsize = 10)
X = gdp_stage['Value']

Y = industry_value['Value']
axis.scatter(X,Y)
plt.show()
```
